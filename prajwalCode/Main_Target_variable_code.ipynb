```python
import time
import codeforces_api
import pandas as pd

parser = codeforces_api.CodeforcesParser()
cf_api = codeforces_api.CodeforcesApi()


def main():

    test = cf_api.problemset_recent_status(1000, problemset_name="")

    id_list = []
    problem_name_list = []
    tags_list = []
    user_list = []
    problemrating_list = []
    user_rating_list = []
    problemindex_list = []
    imp = []
    math = []
    greedy = []
    dp = []
    bit = []
    con = []
    sot = []
    ds = []
    bt = []
    grp = []
    bs = []
    dfs = []
    tree = []
    str = []
    nt = []
    com = []
    geo = []
    tp = []
    dsu = []
    sp = []
    prb = []
    dc = []
    hs = []
    games = []
    fw = []
    it = []
    mat = []
    sss = []
    fft = []
    gm = []
    ts = []
    ep = []
    mm = []
    sat = []
    ct = []
    sch = []

    for i in test:
        ta = (i.problem.tags)
        check = i.author.members
        # print(i.id,':')
        id_list.append(i.id)
        problemrating_list.append(i.problem.rating)
        problem_name = f"{i.problem.contest_id}{i.problem.index}{i.problem.name}"
        problem_name_list.append(problem_name)
        #print(problem_name)
        ta.sort()
        tags_list.append(ta)
        cnt = 0

        til = 0
        for j in check:
            user_list.append(j.handle)
            til+=1
            if(til == 1):
                break

    try:
        users = cf_api.user_info(user_list)
    except:
        cnt += 1
    for i in users:
        try:
            user_rating_list.append(i.rating)
        except:
            user_rating_list.append(0)
    
    a = ({'Submission_Id': id_list, 'Problem Name': problem_name_list, 'User_Rating': user_rating_list, 'User_Name': user_list, 'Tags': tags_list, 'Problem_Rating': problemrating_list, 
#           'Implementation': imp, 'Maths': math, 'Greedy': greedy, 'dp': dp,'Sorting': sot, 'Bitmasks': bit, 'Constructive Algo': con, 'Data Structure': ds, 'Brute force': bt, 'Graph': grp, 'Binary Search': bs, 'dfs and similar': dfs, 'Trees': tree, 'String': str, 'Number Theory': nt, 'Combinatorics': com, 'Geometry': geo, 'Two Pointers': tp, 'dsu': dsu, 'Shortest Paths': sp, 'Probabilities': prb, 'Divide and Conquer': dc, 'Hashing': hs, 'Games': games, 'Flows': fw, 'Interactive': it, 'Matrices': mat,
#           'String Suffix Structures': sss, 'fft': fft, 'Graph Matchings': gm, 'Ternary Search': ts, 'Expression Parsing': ep, 'Meet-in-the-Middle': mm, '2-sat': sat, 'Chinese Remainder Theorem': ct, 'Schedules': sch
         })

    df = pd.DataFrame.from_dict(a, orient='index')

    df = df.transpose()

    df.to_csv('prob_rec.csv',
              mode='a', index=False)
    print("done!!")          


# compression_opts = dict(method='zip',
#                         archive_name='ProbRecco.csv')
# df.to_csv('IEProject.zip', index=False,
#           compression=compression_opts)


if __name__ == "__main__":
    for i in range(20):
        main()
        time.sleep(10*50)
        print(i, "th sleep this is")
```
